```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import unittest
from scipy.stats import linregress
```

```python
def draw_plot():
    # Read data from file
    df = pd.read_csv('epa-sea-level.csv')

    # Create scatter plot
    x1 = df['Year']
    y1 = df['CSIRO Adjusted Sea Level']
    x2 = df[df['Year'] >= 2000]['Year']
    y2 = df[df['Year'] >= 2000]['CSIRO Adjusted Sea Level']

    fig, ax = plt.subplots()
    plt.scatter(x1, y1, color = 'hotpink')

    # Create first line of best fit
    slope1, intercept1, r_value, p_value, std_err  = linregress(x1, y1)
    regression_line_x1 = [x1.min(), 2050]
    regression_line_y1 = [(lambda x: slope1 * x + intercept1)(x) for x in regression_line_x1]
    ax.plot(regression_line_x1, regression_line_y1, color='red')

    # Create second line of best fit
    slope2, intercept2, r_value, p_value, std_err  = linregress(x2, y2)
    regression_line_x2 = [x2.min(), 2050]
    regression_line_y2 = [(lambda x: slope2 * x + intercept2)(x) for x in regression_line_x2]
    ax.plot(regression_line_x2, regression_line_y2, color='blue')

    # Add labels and title
    ax.set(xlabel="Year", ylabel="Sea Level (inches)", title="Rise in Sea Level")
    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```
